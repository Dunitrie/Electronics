```python
import pandas as pd
import matplotlib.pyplot as plt
dateipfad = '5_2_j.csv'
df = pd.read_csv(dateipfad)


plt.plot(df['Time (s)'], df['Vin'], label='Vin')
plt.plot(df['Time (s)'], df['Inv1'], label='Inv1')
plt.plot(df['Time (s)'], df['Inv2'], label='Inv2')
plt.plot(df['Time (s)'], df['Inv3'], label='Inv3')

plt.xlabel('Time (s)')
plt.ylabel('Values')
plt.legend()
plt.show()
```
