```python
import os 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pathlib
```

```python
csv_base_path = ".\Csv_files"
plots_save_path = ".\Plots_Simulation"
```

3.2 MOSFET characteristics

 <br>



```python
csv_file = os.path.join(csv_base_path, "3_2_b.txt")
print(csv_file)
current_data = []

with open(csv_file, 'r') as file:
    for line_idx, line in enumerate(file):
        if line_idx == 0:
            cols = line.strip().split("\t")
        else: 
            if line.strip().startswith("Step") or line.strip().startswith('v'):
                if current_data:
                    df = pd.DataFrame(current_data, columns=cols)
                    dfs.append(df)
                    current_data = []
            else:
                current_data.append([float(val) for val in line.strip().split(sep="\t")])
df = pd.DataFrame(current_data, columns=cols)
print(df.head())

fig, ax1 = plt.subplots()
ax2 = ax1.twinx()

ax1.plot(df.loc[:, "V(vgs)"], df.loc[:,"V(vo)"], "r")
ax1.set_ylabel("$V_{o}$[V]", color="r")

ax2.plot(df.loc[:, "V(vgs)"], df.loc[:,"I(Rd)"]*10**3, "b")
ax2.set_xlabel("$V_{GS}$[V]")
ax2.set_ylabel("$I_D$[mA]", color="b")


plt.savefig(os.path.join(plots_save_path, "3_2_1_b.png"))
plt.show()
```

```python
csv_file = os.path.join(csv_base_path, "3_2_h.txt")
print(csv_file)
current_data = []
dfs = []

with open(csv_file, 'r') as file:
    for line_idx, line in enumerate(file):
        if line_idx == 0:
            cols = line.strip().split("\t")
        else: 
            if line.strip().startswith("Step") or line.strip().startswith('v'):
                if current_data:
                    df = pd.DataFrame(current_data, columns=cols)
                    dfs.append(df)
                    current_data = []
            else:
                current_data.append([float(val) for val in line.strip().split(sep="\t")])
df = pd.DataFrame(current_data, columns=cols)
print(df)
dfs.append(df)

points = np.zeros((3,2))
legends = [f"$V_i$={val}V" for val in [2.19, 2.23, 2.255]]
fig, ax = plt.subplots()
for df_idx, df in enumerate(dfs):
    ax.plot(df.loc[:, "V(vo)"], df.loc[:,"I(Rd)"])
    points[df_idx]= np.array((df.iloc[-1, 2], df.iloc[-1, 3]))
ax.set_xlabel("$V_{DS}$[V]")
ax.set_ylabel("$I_D$[mA]")

x_opt = points[0,0]-points[0,1]*(points[1,0]-points[0,0])/(points[1,1]-points[0,1])
y_opt = points[0,1]-points[0,0]*(points[1,1]-points[0,1])/(points[1,0]-points[0,0])

ax.plot([0, x_opt], [y_opt, 0])
# ax.text(.5, y_opt, f"(0, {y_opt:.1f})")
# ax.text(x_opt, .5, f"({x_opt:.1f})") 
print(f"X_opt={x_opt}, Y_opt={y_opt}")
plt.legend(legends)
plt.grid()
plt.savefig(os.path.join(plots_save_path, "3_2_i.png"))
plt.show()
```

```python
csv_file = os.path.join(csv_base_path, "3_3_1_e.txt")
print(csv_file)
current_data = []
dfs = []

with open(csv_file, 'r') as file:
    for line_idx, line in enumerate(file):
        if line_idx == 0:
            cols = line.strip().split("\t")
        else: 
            if line.strip().startswith("Step") or line.strip().startswith('v'):
                if current_data:
                    df = pd.DataFrame(current_data, columns=cols)
                    dfs.append(df)
                    current_data = []
            else:
                current_data.append([float(val) for val in line.strip().split(sep="\t")])
df = pd.DataFrame(current_data, columns=cols)
print(df.head())
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()

ax1.plot(df.loc[:, "time"], df.loc[:, "V(vgs)"], "r")
ax1.set_ylabel("$V_{GS}$[V]", color="r")

ax2.plot(df.loc[:, "time"], df.loc[:, "V(vo)"], "b")
ax1.set_xlabel("t[s]")
ax2.set_ylabel("$V_{o}$[V]", color="b")

plt.savefig(os.path.join(plots_save_path, "3_3_1_e.png"))
plt.show()
print(f"Max_voltage $V_G$={np.max(df.loc[:,'V(vgs)'])}")
print(f"Max_voltage $V_D$={np.max(df.loc[:,'V(vo)'])}")
```

# 3.4 Step Up Converter

 <br>



```python
csv_files = ["3_4_d_5.txt",
             "3_4_d_20.txt",
             "3_4_d_50.txt",
             "3_4_d_60.txt"]
legends = [f"{val}%" for val in [5, 20, 50, 60]]
csv_files = [os.path.join(csv_base_path, csv_file) for csv_file in csv_files]
colors = ["blue", "orange", "green", "red"]
             
dfs = []
current_data = []
fig, ax = plt.subplots()
for csv_idx, csv_file in enumerate(csv_files):
    with open(csv_file, 'r') as file:
        for line_idx, line in enumerate(file):
            if line_idx == 0:
                cols = line.strip().split("\t")
            else: 
                if line.strip().startswith("Step") or line.strip().startswith('v'):
                    if current_data:
                        df = pd.DataFrame(current_data, columns=cols)
                        dfs.append(df)
                        current_data = []
                else:
                    current_data.append([float(val) for val in line.strip().split(sep="\t")])
    df = pd.DataFrame(current_data, columns=cols)
    print(df.head())
    ax.plot(df.loc[:, "time"]*1000, df.loc[:,"V(n002)"], colors[csv_idx])
    dfs.append(df)

ax.set_xlabel("t[ms]")
ax.set_ylabel(r"V_{dc-out}[V]")
plt.legend(legends)

plt.savefig(os.path.join(plots_save_path, "3_4_d.png"))
# plt.plot(df.loc[:, "V(vbe)"], -df.loc[:,"I(Rb)"]*10**6)
# plt.xlabel("$V_{BE}$[V]")
# plt.ylabel("$I_B[\mu$A]")
# plt.savefig(os.path.join(plots_save_path, "2_2_1_e.png"))
# plt.show()

# legends = ["$I_C$[mA]", r"$\beta$[1]"]
# fig, ax1 = plt.subplots()
# ax2 = ax1.twinx()


# ax1.plot(-df.iloc[100:, 2]*1000, df.iloc[100:,3]*1000, 'r')
# ax1.set_ylabel(legends[0], color='r')

# ax2.plot(-df.iloc[100:, 2]*1000, df.iloc[100:,3]/-df.iloc[100:,2], 'b')
# ax2.set_ylabel(legends[1], color='b')
# ax1.set_xlabel("$I_B$[mA]")
# plt.savefig(os.path.join(plots_save_path, "2_2_1_f.png"))
# plt.show()
```

```python
csv_file = csv_files[1]
print(csv_file)
current_data = []
dfs = []
current_data = []

with open(csv_file, 'r') as file:
    for line_idx, line in enumerate(file):
        if line_idx == 0:
            cols = line.strip().split("\t")
        else: 
            if line.strip().startswith("Step") or line.strip().startswith('v'):
                if current_data:
                    df = pd.DataFrame(current_data, columns=cols)
                    dfs.append(df)
                    current_data = []
            else:
                current_data.append([float(val) for val in line.strip().split(sep="\t")])
df = pd.DataFrame(current_data, columns=cols)
dfs.append(df)
print(df)
```

```python
legends = [f"$I_B$={val}"+r"$\mu A$" for val in [0, 4, 13, 23, 33, 43]]
for df_idx, df in enumerate(dfs):
    plt.plot(df.loc[:, "V(vce)"], df.loc[:,"I(Rc)"]*10**3)
plt.xlabel("$V_{CE}$[V]")
plt.ylabel("$I_C[\mu$A]")
plt.legend(legends)

for idx in range(len(dfs)):
    df = dfs[idx]
    x1 = 500
    vce1 = df.iloc[x1,1]
    ic1 = df.iloc[x1, 2]*1e3

    x2 = 800
    vce2 = df.iloc[x2,1]
    ic2 = df.iloc[x2, 2]*1e3

    d = (ic2-ic1)/(vce2-vce1)
    early_voltage = vce1-ic1/d
    print(early_voltage)
    # print(x2, x1)
    # print(y2, y1)
    # print(vce2, vce1)
    plt.scatter([vce1, vce2], [ic1, ic2])


#plt.savefig(os.path.join(plots_save_path, "2_2_1_h.png"))
plt.show()
```
